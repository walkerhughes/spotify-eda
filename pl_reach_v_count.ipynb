```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import plot_settings
```

```python
dfs = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='latin1')
dfs.head()
```

```python
dfs.info()
```

```python
# Much of the data are 'object' when they should be 'int'
dfs['Spotify Streams'] = dfs['Spotify Streams'].str.replace(',', '').astype(float)
dfs['Spotify Playlist Count'] = dfs['Spotify Playlist Count'].str.replace(',', '').astype(float)
dfs['Spotify Playlist Reach'] = dfs['Spotify Playlist Reach'].str.replace(',', '').astype(float)
```

```python
# Drop all data points that are 'NaN'.
dfs1 = dfs.dropna(subset=['Spotify Streams','Spotify Playlist Count','Spotify Playlist Reach'])
```

```python
#I want to maintain the name dfs 
dfs = dfs1
```

```python
x = dfs['Spotify Playlist Count']
y = dfs['Spotify Playlist Reach']

r = np.corrcoef(x,y)[1][0]
a = r*(y.std()/x.std())
b = y.mean() - (a*x.mean())

x_lin = np.linspace(min(x), max(x), 10)
y_lin = a*x_lin + b
```

```python
np.corrcoef(x,y)
```

```python
fig, ax = plt.subplots(figsize=(9,4))

# Set a color scheme such that default color is #1E90FF (skyblue) and #E15F32 (orange) for "premium playlists" (reach > 150M)
col = np.where(y<1.5e8,'#1E90FF','#E15F32')
# Make the premiums visible with a larger size and thicker color
ss = np.where(y<1.5e8,1,5)
alp = np.where(y<1.5e8,0.5,1)

ax.scatter(x, y, c=col, s=ss, alpha=alp)
ax.plot(x_lin, y_lin, c='red', lw=0.8, alpha=0.8)

ax.annotate(f'r = {r.round(3)}\ny = {int(a.round())}*x + {b:.2e}',(4.5e5,1.8e8),c='red',size=10)
ax.annotate("Premium Songs (150M+ Reach)",(.0e5,2.3e8),c='#E15F32',size=10, weight='bold')

ax.set_title("Spotify's Playlist Count vs Reach", pad=8, size=15, weight='bold')
ax.set_xlabel("Playlist Count by Song", labelpad=5, size=11)
ax.set_ylabel("# of Followers of All Playlists by Song", labelpad=5, size=11)

plt.show()
```

```python
# 0.5 < r < 0.7 indicates a moderate correlation between playlist count and reach.

# This data displays a number of songs that have a high reach despite having a low count.

# "Premium Playlists" have been highlighted in orange (reach > 150M) which will get recommended to paid users.
```

```python
# Playlist count vs reach by weighing in streams

quantiles = dfs['Spotify Streams'].quantile([0.25, 0.75])
q_low = quantiles[0.25]
q_high = quantiles[0.75]

group1 = dfs[dfs['Spotify Streams'] <= q_low]
group2 = dfs[(dfs['Spotify Streams'] > q_low) & (dfs['Spotify Streams'] <= q_high)]
group3 = dfs[dfs['Spotify Streams'] > q_high]
```

```python
x1 = group1['Spotify Playlist Count']
y1 = group1['Spotify Playlist Reach']
x2 = group2['Spotify Playlist Count']
y2 = group2['Spotify Playlist Reach']
x3 = group3['Spotify Playlist Count']
y3 = group3['Spotify Playlist Reach']

r1 = np.corrcoef(x1,y1)[1][0]
a1 = r1*(y1.std()/x1.std())
b1 = y1.mean() - (a1*x1.mean())

r2 = np.corrcoef(x2,y2)[1][0]
a2 = r2*(y2.std()/x2.std())
b2 = y2.mean() - (a2*x2.mean())

r3 = np.corrcoef(x3,y3)[1][0]
a3 = r3*(y3.std()/x3.std())
b3 = y3.mean() - (a3*x3.mean())

x1_lin = np.linspace(min(x), max(x), 10)
y1_lin = a1*x1_lin + b1

x2_lin = np.linspace(min(x), max(x), 10)
y2_lin = a2*x2_lin + b2

x3_lin = np.linspace(min(x), max(x), 10)
y3_lin = a3*x3_lin + b3
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(9,4))

# Set equal size axes for all 3 plots
for i in range(len(axes)): 
    axes[i].set_xlim(-30000, 640000)
    axes[i].set_ylim(-.2e8, 2.8e8)

# Scatter plots and regression lines
axes[0].scatter(x1,y1,c='#86a2cf', s=5, alpha=0.5)
axes[1].scatter(x2,y2,c='#1E90FF', s=5, alpha=0.5)
axes[2].scatter(x3,y3,c='#1AA260', s=5, alpha=0.5)
axes[0].plot(x1_lin, y1_lin, c='red', lw=1, alpha=0.8)
axes[1].plot(x2_lin, y2_lin, c='red', lw=1, alpha=0.8)
axes[2].plot(x3_lin, y3_lin, c='red', lw=1, alpha=0.8)

# Annotate the linear regressions
axes[0].annotate(f'r = {r1.round(3)}\ny = {int(a1.round())}*x + {b1:.2e}',(1.2e5,.3e8),c='red',size=10) 
axes[1].annotate(f'r = {r2.round(3)}\ny = {int(a2.round())}*x + {b2:.2e}',(1.4e5,0.95e8),c='red',size=10) 
axes[2].annotate(f'r = {r3.round(3)}\ny = {int(a3.round())}*x + {b3:.2e}',(1.1e5,1.9e8),c='red',size=10) 
axes[0].annotate("Little to No Correlation",(.165e5,2.45e8), c='#E15F32', size=10, weight='bold')
axes[1].annotate("Little to No Correlation",(.165e5,2.45e8), c='#E15F32', size=10, weight='bold')
axes[2].annotate("Moderate Correlation",(.35e5,2.48e8), c='#E15F32', size=10, weight='bold')

# Show the titles
axes[1].text(-7.3e5,3.3e8,"Spotify's Playlist Count vs Reach by Streams",weight='bold',size=18)
axes[0].set_title("Low Stream (0%,25%)", pad=17, size=10)
axes[1].set_title("Medium Stream (25%,75%)", pad=17, size=10)
axes[2].set_title("High Stream (75%,100%)", pad=17, size=10)
axes[1].set_xlabel("Playlist Count per Each Song", labelpad=10, size=14)
axes[0].set_ylabel("# Followers of Song's Playlists", labelpad=10, size=14)

plt.show()
```
